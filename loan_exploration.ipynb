```python
plt.figure(figsize=(15,5))

#order the category differently to the way it is in the dataframe so the past due days are ordered
level_order = ['Current', 'Completed', 'Chargedoff','FinalPaymentInProgress','Defaulted','Past Due (1-15 days)',
               'Past Due (16-30 days)','Past Due (31-60 days)','Past Due (61-90 days)','Past Due (91-120 days)']
ordered_cat = pd.api.types.CategoricalDtype(ordered = True, categories = level_order)
loan_data['LoanStatus'] = loan_data['LoanStatus'].astype(ordered_cat)

# plot the graph and set the bars to one colour
base_color = sb.color_palette()[0]
l_plot = sb.countplot(data = loan_data, y = 'LoanStatus', color = base_color)

# place frequency values of each loan status category in the graph
for p in l_plot.patches:
        percentage = '{:.0f}'.format(p.get_width())
        x = p.get_x() + p.get_width() + 0.02
        y = p.get_y() + p.get_height()/2
        l_plot.annotate(percentage, (x, y))
```

```python
plt.figure(figsize = [15, 8]) # larger figure size for subplots

# create sample of data so plots render faster
samples = np.random.choice(loan_data.shape[0], 5000, replace = False)
loan_sample = loan_data.iloc[samples,:]

# monthly income
plt.subplot(2, 2, 1)
binsize= 3000
bins = np.arange(0, loan_sample['StatedMonthlyIncome'].max()+binsize, binsize)
plt.hist(data=loan_sample, x= 'StatedMonthlyIncome',bins=bins)
plt.xlabel('monthly income')

# monthly repayments
plt.subplot(2, 2, 2)
binsize=50
bins = np.arange(0, loan_sample['MonthlyLoanPayment'].max()+binsize, binsize)
plt.hist(data=loan_sample, x= 'MonthlyLoanPayment',bins=bins)
plt.xlabel('monthly loan payment')

#original loan amount
plt.subplot(2, 2, 3)
binsize=500
bins = np.arange(0, loan_sample['LoanOriginalAmount'].max()+binsize, binsize)
plt.hist(data=loan_sample, x= 'LoanOriginalAmount',bins=bins)
plt.xlabel('original loan amount')

#months since loan received
plt.subplot(2, 2, 4)
binsize=2
bins = np.arange(0, loan_sample['LoanMonthsSinceOrigination'].max()+binsize, binsize)
plt.hist(data=loan_sample, x= 'LoanMonthsSinceOrigination',bins=bins)
plt.xlabel('number of months since loan was received');
```

> Two of the variables, namely monthly income and monthly repayment are skewed to the right. One cannot use logarithmic transformation of these variables as there are many zero values so a square root transformation can be done, which will be done next.

```python
plt.figure(figsize = [15, 5])

# square root transformation
def sqrt_trans(x, inverse = False):
    """ transformation helper function """
    if not inverse:
        return np.sqrt(x)
    else:
        return x ** 2


# monthly income subplot
plt.subplot(1, 2, 1)
bin_edges = np.arange(0, sqrt_trans(loan_sample['StatedMonthlyIncome'].max())+5, 5)
plt.hist(loan_sample['StatedMonthlyIncome'].apply(sqrt_trans), bins = bin_edges)
tick_locs = np.arange(0, sqrt_trans(loan_sample['StatedMonthlyIncome'].max())+50, 50)
plt.xticks(tick_locs, sqrt_trans(tick_locs, inverse = True).astype(int), rotation=25);
plt.xlabel('monthly income');

# monthly repayments subplot
plt.subplot(1, 2, 2)
bin_edges = np.arange(0, sqrt_trans(loan_sample['MonthlyLoanPayment'].max())+1, 1)
plt.hist(loan_sample['MonthlyLoanPayment'].apply(sqrt_trans), bins = bin_edges)
tick_locs = np.arange(0, sqrt_trans(loan_sample['MonthlyLoanPayment'].max())+10, 10)
plt.xticks(tick_locs, sqrt_trans(tick_locs, inverse = True).astype(int));
plt.xlabel('monthly loan payment');
```

> Once the square root transformation has been done, the monthly income variable distribution is unimodal and looks like most customers having an average monthly income of around 5000. The monthly repayment variable seems a little bit bimodal with one peak around the 200 mark and another around 400.
Thus doing a slight comparison, it would seem on average with a monthly income of 5000, 200 to 400 of it is for a loan payment.

> Lastly, the relationship between loan status, interest rate and original loan amount will be looked at.

```python
#create function for frequency polygons for levels of loan status
def mean_poly(x, y, bins = 10, **kwargs):
    """ Custom adapted line plot code. """
    # set bin edges if none or int specified
    if type(bins) == int:
        bins = np.linspace(x.min(), x.max(), bins+1)
    bin_centers = (bin_edges[1:] + bin_edges[:-1]) / 2

    # compute counts
    data_bins = pd.cut(x, bins, right = False,
                       include_lowest = True)
    means = y.groupby(data_bins).mean()

    # create plot
    plt.errorbar(x = bin_centers, y = means, **kwargs)

bin_edges = np.arange(0.25, loan_data['BorrowerRate'].max()+0.05, 0.05)
g = sb.FacetGrid(data = loan_data, hue = 'LoanStatus', height = 7)
g.map(mean_poly, "BorrowerRate", "LoanOriginalAmount", bins = bin_edges)
g.set_ylabels('mean(LoanOriginalAmount)')
g.add_legend();
```

> The negative relationship between interest rate and average original loan amount seen here was also seen earlier and we can see here that it is negative for all loan status categories.
> Completed, Chargedoff and defaulted loans have interest rates that are higher than 0.33 with the other loan status categories not going beyond 0.33. These three also have less steeper slopes which means that the lower the original loan amounts are, the less rate at which the interest rates increases compared to the rest of the categories.
